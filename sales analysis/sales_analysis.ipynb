```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python


data = {
"order_id": [101,102,103,104,105,106],
"city": ["Delhi","Mumbai","Delhi","Pune","Mumbai","Delhi"],
"product": ["Laptop","Phone","Tablet","Laptop","Phone","Tablet"],
"sales": [50000,30000,20000,45000,35000,25000],
"month": ["Jan","Jan","Feb","Feb","Mar","Mar"]
}

df = pd.DataFrame(data)

```

```python
df
```

Total sales

```python
print(df["sales"].sum())
```

Total sales per city


```python
print(df.groupby("city")["sales"].sum())
```

Sales per product

```python
df.groupby("product")["sales"].sum()
```

sales per month

```python
df.groupby("month")["sales"].sum()
```

```python
df.groupby('city')["sales"].sum().plot(kind="bar")
plt.title("Total sales per city")
plt.xlabel("City")
plt.ylabel("Sales")
```

```python
df.groupby('month')["sales"].sum().plot(kind="line")
plt.title("Total sales per month")
plt.xlabel("Month")
plt.ylabel("Sales")
```

```python
sns.barplot(x="product",y="sales",data=df)
```
